##### How many people have heart disease and how many people don't have heart disease

```python
data.columns
```

```python
data['target'].value_counts()
```

```python
sns.countplot(x = data['target'], data = data)

plt.xlabel("target")
plt.ylabel("Counts")
plt.title("How many people have heart disease and how many people don't have heart disease")


plt.show()
```

##### Find the count of male and female in the dataset

```python
data.columns
```

```python
data['sex'].value_counts()
```

```python
sns.countplot(x = data['sex'])

plt.title("Count of Male and Female")

plt.xticks(ticks=[0, 1], labels=['Female', 'Male'])
plt.show()
```

##### Find the distribution of Gender according to target variable

```python
sns.countplot(x = 'sex',hue = 'target',data = data)

plt.xlabel("sex")
plt.ylabel("Counts")
plt.title("Distribution of Gender according to Target variable")

plt.xticks(ticks=[0, 1], labels=['Female', 'Male'])
plt.legend(labels = ['No Disease','Disease'])
plt.show()
```

##### Check age distribution in the dataset

```python
data.columns
```

```python
sns.histplot(x = 'age',data = data,bins = 20)
plt.title("Distribution of Age")
plt.show()
```

##### Check chest pain type

```python
data.columns
```

```python
sns.countplot(x = 'cp',data = data)

plt.xlabel("cp")
plt.ylabel("Counts")
plt.title("Types of Chest Pain")

plt.xticks(ticks = [0,1,2,3], labels = ["typical angina","atypical angina","non-anginal pain","asymptomatic"])
plt.show()
```

##### Show the chest pain distribution as per target variable

```python
sns.countplot(x = 'cp', hue = 'target', data = data)

plt.xlabel("cp")
plt.ylabel("Counts")
plt.title("Chest Pain Distribution as per Target Variable")

plt.xticks(ticks = [0,1,2,3], labels = ["typical angina","atypical angina","non-anginal pain","asymptomatic"])
plt.legend(labels = ['No Disease','Disease'])
plt.show()
```
